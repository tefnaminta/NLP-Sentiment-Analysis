```python
from sklearn.feature_extraction.text import CountVectorizer
bow1=CountVectorizer().fit_transform(iph['Review Title']).toarray()
bow2=CountVectorizer().fit_transform(iph['Review Body']).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bow1,iph['Rating'],test_size=0.5)
from sklearn.naive_bayes import MultinomialNB
mnclassifier = MultinomialNB().fit(X_train, y_train)
xtrain,xtest,ytrain,ytest= train_test_split(bow2,iph['Rating'],test_size=0.5)
mnclassifier1 = MultinomialNB().fit(xtrain, ytrain)
```

```python
predict = mnclassifier.predict(bow1)
predict1 = mnclassifier1.predict(bow2)
```

```python
iphrate = pd.DataFrame()
iphrate['title'] = predict
iphrate['body'] = predict1
```

```python
iphrate['avg'] = ""
```

```python
for i in range(len(iphrate['title'])):
  iphrate['avg'][i]=round((int(iphrate['title'][i])+int(iphrate['body'][i]))/2)
```
