```python
import pandas as pd
```

```python
iphone = pd.read_csv('iphone_new_rating_data.csv')
```

```python
import re
sentences_pos = []
for i in range(len(iphone['Review Body'])):
  if(iphone['Rating'][i]==5.0):
    sentences_pos.append(re.split(' ',str(iphone['Review Body'][i])))
```

```python
import re
sentences_neg = []
for i in range(len(iphone['Review Body'])):
  if(iphone['Rating'][i]<=2.0):
    sentences_neg.append(re.split(' ',str(iphone['Review Body'][i])))
```

```python
import nltk
from gensim.models import Word2Vec
```

```python
modelpos = Word2Vec(sentences_pos, min_count=10) 
print(modelpos)
```

```python
modelneg = Word2Vec(sentences_neg, min_count=3) 
print(modelneg)
```

```python
icloud_positive = ""
words = modelpos.wv.vocab
for i in words:
  if(len(i)>2):
    icloud_positive=icloud_positive+i+" "
```

```python
icloud_negative = ""
words = modelneg.wv.vocab
for i in words:
  if(len(i)>2):
    icloud_negative=icloud_negative+i+" "
```

```python
samsung = pd.read_csv('samsung_new_rating_data.csv')
```

```python
import re
pos_sentences1 = []
for i in range(len(samsung['Review Body'])):
  if(samsung['Rating'][i]==5.0):
    pos_sentences1.append(re.split(' ',str(samsung['Review Body'][i])))
neg_sentences1 = []
for i in range(len(samsung['Review Body'])):
  if(samsung['Rating'][i]<=2.0):
    neg_sentences1.append(re.split(' ',str(samsung['Review Body'][i])))
```

```python
pos_model1 = Word2Vec(pos_sentences1, min_count=7) 
print(pos_model1)
neg_model1 = Word2Vec(neg_sentences1, min_count=5) 
print(neg_model1)
```

```python
scloud_pos = ""
pos_words1 = pos_model1.wv.vocab
for i in pos_words1:
  if(len(i)>2):
    scloud_pos = scloud_pos+i+" "
scloud_neg = ""
neg_model1 = neg_model1.wv.vocab
for i in neg_model1:
  if(len(i)>2):
    scloud_neg = scloud_neg+i+" "
```
