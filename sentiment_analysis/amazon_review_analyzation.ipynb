```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
iph = pd.read_csv('iphone_new_rating_data.csv')
```

```python
iphone = pd.read_csv('amazon_review.csv')
iphone = iphone.drop('Unnamed: 0',axis=1)
iphone['New Rating'] = iph['New Rating']
iphone
```

```python
iphone.corr()
```

```python
sb.heatmap(iphone.corr())
```

```python
rating_count = pd.DataFrame()
rating_count['Date'] = iphone['Date'].unique()
rating_count['Count'] = [i for i in iphone['Date'].value_counts()]
rating_count.head()
```

```python
rating_compare = pd.DataFrame()
rating_compare['Actual Rating'] = iphone['Rating']
rating_compare['Rating by Review'] = iphone['New Rating']
ss = []
for i in range(len(iphone['New Rating'])):
  if(iphone['Rating'][i]>iphone['New Rating'][i]):
    ss.append(round(((iphone['Rating'][i]-iphone['New Rating'][i])/iphone['Rating'][i])*100))
  else:
    ss.append(round(((iphone['New Rating'][i]-iphone['Rating'][i])/iphone['New Rating'][i])*100))
rating_compare['Difference in %'] = ss

ss = []
for i in rating_compare['Difference in %']:
  if(i==100):
    ss.append('High Difference')
  elif(i==0):
    ss.append('No Difference')
  else:
    ss.append('Medium Difference')
rating_compare['Comparison'] = ss
rating_compare
```

```python
rating_compare.to_csv('iphone_rating_comparison.csv')
```
